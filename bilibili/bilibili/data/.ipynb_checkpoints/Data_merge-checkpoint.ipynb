```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('Bilibili_Info-1.csv',encoding='utf-8')
df1=df.drop_duplicates('aid')
```

```python
df1=df1[df1['view'] > df1['view'].min()]
temp=df1.drop_duplicates('owner_mid')
df2 = pd.read_csv('Up_Fan.csv',encoding='utf-8')
df2=pd.merge(df2,temp,on='owner_mid')
#df3为up名，mid和fan数量
df3=df2[['owner_name','fan']]
df3
```

```python

up_top= df1[['view','share','reply','like','coin','favorite','danmaku','videos','duration']].groupby(df1['owner_name']).sum()
up_top=up_top.reset_index()
up_top=pd.merge(up_top,df3,on='owner_name')
up_top=up_top.sort_values(by='view',ascending=False)

#up_top_20=up_top.sort_values(by='view',ascending=False)
#top10=up_top_20[:10]
up_top.to_csv('Up_Influence.csv',index =0)
```
