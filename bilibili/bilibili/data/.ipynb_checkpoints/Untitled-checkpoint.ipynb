```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#数据汇总并去重
#df =pd.read_csv('Bilibili_info-1.csv', encoding='utf-8')
#df.to_csv('Bilibili_Info.csv', mode ="a",header=None,index =0)
df = pd.read_csv('Bilibili_info-1.csv',encoding='utf-8')
df1=df.drop_duplicates('aid')
```

```python
#df1为初始数据
#去除-1的被删除视频
df1=df1[df1['view'] > df1['view'].min()]
df2=df1
```

```python
#view 计算
df2=df2.groupby(by='view').size()
df2=df2.reset_index()
df2
```

```python
df3 = df2
df3
#plt.hist(df2['view'],bins=200)
```

```python
#按照up的总播放量进行排名
df4=df1
up_top= df4[['view','share','reply','like','coin','favorite','danmaku','videos','duration']].groupby(df4['owner_name']).sum()
up_top=up_top.reset_index()
up_top_20=up_top.sort_values(by='view',ascending=False)[:20]
top10=up_top_20[:10]
#输出
top10.to_csv('Top10_Up.csv',index =0)
```

```python
up_top.to_csv('Up.csv',index =0)
```

```python
#OLS模型引用
#share view
import statsmodels.api as sm
import statsmodels.formula.api as smf
from __future__ import print_function
from statsmodels.sandbox.regression.predstd import wls_prediction_std

#sm.add_constant(up_top['reply'])
x=up_top['share']
y=up_top['view']

model = sm.OLS(y, x)
results = model.fit()
results.summary()
y_fitted = results.fittedvalues
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, 'o', label='data')
ax.plot(x, y_fitted, 'r--.',label='OLS')
ax.legend(loc='best')
results.summary()
```
